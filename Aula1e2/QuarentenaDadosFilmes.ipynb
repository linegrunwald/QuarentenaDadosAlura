##Desafio 6
Plotar o boxplot e o histograma um do lado do outro (na mesma figura ou em figuras distintas)

```
def plot_filme(n):
  notas_do_filme = avaliacoes.query(f"filmeId=={n}")["nota"]
  f, axes = plt.subplots(1, 2, figsize=(16, 8))
  notas_do_filme.plot(kind='hist',ax=axes[0])
  notas_do_filme.plot.box(ax=axes[1])
  plt.show()
  return notas_do_filme.describe()
```

#Apenas um teste de função para n plots

```
def plot_filmes(idsFilme):
  colunas = len(idsFilme)
  fig, eixos = plt.subplots(nrows=1, ncols=colunas, figsize=(30,15))
  coluna = 0
  for n in idsFilme:
    notas_do_filme = avaliacoes.query(f"filmeId=={n}")["nota"]
    notas_do_filme.plot(kind='hist',ax=eixos[coluna])
    
    coluna = coluna+1

  plt.show()
```

```
plot_filmes([1,2,3])
```
